```python
dados = dadosJair[['sg_uf', 'nm_municipio', 'qt_votos_nom_validos']]
```

```python
dados.columns = ['estado', 'municipio', 'votos_Bolsonaro']
```

```python
dados['votos_Lula'] = 0
dados
```

```python
for i in range(dadosLula.shape[0]):
    if dados.estado[i] == dadosLula.sg_uf[i]:
        dados.votos_Lula[i] = (dadosLula.qt_votos_nom_validos[i])
    
```

```python
dados = dados[['regiao','estado','municipio','votos_Bolsonaro','votos_Lula']].sort_values(by='municipio')
dados
```

```python
import unicodedata
```

```python

for i in range(len(dados.municipio)):
    dados.municipio[i] = ''.join(ch for ch in unicodedata.normalize('NFKD', dados.municipio[i]) 
        if not unicodedata.combining(ch))
```

```python
dados = dados.sort_values(by='municipio')
```
