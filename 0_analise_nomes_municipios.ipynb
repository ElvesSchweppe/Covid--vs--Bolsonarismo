```python
import pandas as pd
import numpy as np
```

```python
covid = pd.read_csv('dados/SUS/COVIDBR_2023_17jan2023.csv', sep=';')
eleicao18 = pd.read_csv('dados/TSE/2018_Bolsonaro.csv', sep=';')
eleicao22 = pd.read_csv('dados/TSE/2022_Bolsonaro.csv', sep=';')
```

```python
covid = covid[['regiao', 'estado', 'municipio', 'data']]
covid = covid[covid['data'] == '2023-01-17']
covid = covid[~((covid.estado.isna()) | (covid.municipio.isna()))]
covid.reset_index(inplace=True)
covid = covid.drop(columns = ['index', 'data'])
covid['municipio_estado'] = (covid.municipio + ' / ' + covid.estado)
covid
```

```python
import unicodedata
```

```python
for i in range(len(covid.municipio_estado)):
    covid.municipio_estado[i] = ''.join(ch for ch in unicodedata.normalize('NFKD', covid.municipio_estado[i].upper())
        if not unicodedata.combining(ch))

covid
```

```python
eleicao22
```

```python
eleicao18 = eleicao18[['sg_uf','nm_municipio']]
eleicao22 = eleicao22[['sg_uf','nm_municipio']]
eleicao18['municipio_estado'] = eleicao18.nm_municipio + ' / ' + eleicao18.sg_uf
eleicao22['municipio_estado'] = eleicao22.nm_municipio + ' / ' + eleicao22.sg_uf
eleicao18 = eleicao18[eleicao18.sg_uf != 'ZZ']
eleicao22 = eleicao22[eleicao22.sg_uf != 'ZZ']
eleicao18.columns = ['uf_18', 'mun_18', 'mun_est_18']
eleicao22.columns = ['uf_22', 'mun_22', 'mun_est_22']
```

```python
for i in range(len(eleicao18.mun_est_18)):
    eleicao18.mun_est_18[i] = ''.join(ch for ch in unicodedata.normalize('NFKD', eleicao18.mun_est_18[i]).upper() 
        if not unicodedata.combining(ch))


for i in range(len(eleicao22.mun_est_22)):
    eleicao22.mun_est_22[i] = ''.join(ch for ch in unicodedata.normalize('NFKD', eleicao22.mun_est_22[i]).upper() 
        if not unicodedata.combining(ch))

```

```python
covid = covid.sort_values(by='municipio_estado')
covid.reset_index(inplace=True)
covid = covid.drop(columns=['index'])

eleicao18 = eleicao18.sort_values(by='mun_est_18')
eleicao18.reset_index(inplace=True)
eleicao18 = eleicao18.drop(columns=['index'])

eleicao22 = eleicao22.sort_values(by='mun_est_22')
eleicao22.reset_index(inplace=True)
eleicao22 = eleicao22.drop(columns=['index'])
```

```python
dados = pd.DataFrame(columns=['covid', 'eleicao18', 'eleicao22'])
```

```python
dados['covid'] = covid.municipio_estado
dados['eleicao18'] = eleicao18.mun_est_18
dados['eleicao22'] = eleicao22.mun_est_22
```

```python
dados
```

### Optei por levar todos os nomes para o dado mais atual - 'eleicao22'

```python
lista_22 = dados.eleicao22.to_list()
lista_18 = dados.eleicao18.to_list()
lista_cov = dados.covid.to_list()
```

```python
dados['corr_covid'] = (~dados['covid'].isin(lista_22)).astype(int)
dados['22_covid'] = (~dados['eleicao22'].isin(lista_cov)).astype(int)

dados['corr_18'] = (~dados['eleicao18'].isin(lista_22)).astype(int)
dados['22_18'] = (~dados['eleicao22'].isin(lista_18)).astype(int)

dados['covid_18'] = (~dados['eleicao18'].isin(lista_cov)).astype(int)
dados['18_covid'] = (~dados['covid'].isin(lista_18)).astype(int)

```

```python
dados['corrigir'] = dados['corr_covid'] + dados['22_covid'] + dados['corr_18'] + dados['22_18'] + dados['covid_18'] + dados['18_covid']
```
